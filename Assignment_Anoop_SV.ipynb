```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to python environment.


```python
iris=pd.read_excel("iris.xls")
iris
```

# 2. columns in the dataset


```python
iris.columns
```

# 3. mean of each column of the dataset


```python
#calculating the mean of each columns and the value is assigned to a variable

slmean=iris['SL'].mean()
swmean=iris['SW'].mean()
plmean=iris['PL'].mean()
pwmean=iris['PW'].mean()

#mean of each column is printed.

print('mean of SL: ',slmean)
print('mean of SW: ',swmean)
print('mean of PL: ',plmean)
print('mean of PW: ',pwmean)
```

# 4. null values present in the dataset


```python
#sum of null values
iris.isna().sum()
```

# 5. Visualizations


```python
#scatterplot of petal width vs petal length of different classifications using seaborn
plt.figure(figsize=(10,5)) #to increase the width and height
sns.scatterplot(x=iris['PW'], y=iris['PL'], hue=iris['Classification'])
plt.title('Petal width VS Petal Length ')
plt.xlabel('Petal Width (PL)')
plt.ylabel('Petal Length (PW)')
plt.show()        
```

```python
#boxplot of sepal length of different classifications using seaborn
plt.figure(figsize=(10,5))
sns.boxplot(x=iris['Classification'],y=iris['SL'])
plt.title('Sepal length of different classification')
plt.ylabel('Sepal Length (SL)')
plt.show()
```

```python
#pair plot of sepal length and sepal width using seaborn
sns.pairplot(iris[['SL','SW']])
plt.show()
```

```python
#violinplot of sepal width of each classifications
plt.figure(figsize=(10,5))
sns.violinplot(x=iris['Classification'] , y=iris['SW'])
plt.ylabel('sepal width (SW)')
plt.show()
```

```python
#count of each classifications 
sns.countplot(iris['Classification'])
plt.show()
```

```python
#line plot of Petal length and petal width of each classifications.
plt.figure(figsize=(8,4))
plt.plot(iris['Classification'],iris['PW'],c='red',label='Petal width',linewidth=3)
plt.plot(iris['Classification'],iris['PL'],c='blue',label='petal length',linewidth=3)
plt.xlabel('Classification')
plt.ylabel('Petal width and Petal length ')
plt.ylim(0,8)       #to set  the y axis limits
plt.title('Petal length and petal width of each classifications')
plt.legend()                 
plt.show()
```

```python
#bar plot of petal length of each classification using matplotlib
plt.bar(iris['Classification'], iris['PL'])
plt.title('Petal length of each classifications')
plt.xlabel('Classification')
plt.ylabel('petal length (PL)')
plt.show()
```
